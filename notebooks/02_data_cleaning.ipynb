```python
import pandas as pd

df = pd.read_csv("../data/raw/survey.csv")

df.head()
```

```python
df = df[(df['Age'] >= 15) & (df['Age'] <= 75)]
df.shape
```

```python
def clean_gender(gender):
    gender = str(gender).lower().strip()

    male_terms = ['male', 'm', 'man', 'cis male', 'male-ish', 'maile']
    female_terms = ['female', 'f', 'woman', 'cis female', 'femake']

    if gender in male_terms:
        return 'Male'
    elif gender in female_terms:
        return 'Female'
    else:
        return 'Other'

df['Gender'] = df['Gender'].apply(clean_gender)
```

```python
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.to_csv("../data/processed/cleaned_survey.csv", index=False)
```
