```python
import pandas as pd

df = pd.read_csv("../data/raw/survey.csv")

df.shape
df.head()
df.info()
df.isnull().sum()
df['treatment'].value_counts()
df.describe()
```

The target variable is treatment. Since the model predicts Yes or No, this is a binary classification task. Accuracy alone is not enough because if the data is imbalanced, the model may predict the majority class well but fail to detect the minority class. Therefore, F1 score, precision, recall, and ROC-AUC are more meaningful.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x='treatment')
plt.title("Treatment Class Distribution")
plt.show()
```
